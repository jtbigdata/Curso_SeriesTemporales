#### 1. Introducción a las Series Temporales

Las series temporales son conjuntos de datos recogidos a intervalos regulares a lo largo del tiempo. Este tipo de análisis es esencial en diversas disciplinas como la economía, la meteorología y la ingeniería, ya que permite identificar patrones, tendencias y ciclos en los datos.

Los componentes principales de una serie temporal son:

- **Tendencia**: La dirección general de la serie a largo plazo. Se puede calcular mediante una regresión lineal:

  $
  y_t = \beta_0 + \beta_1 t + \epsilon_t
  $

  donde $y_t$ es el valor en el tiempo $t$, $\beta_0$ es la intersección, $\beta_1$ es la pendiente (tendencia) y $\epsilon_t$ es el término de error.

- **Estacionalidad**: Patrones que se repiten a intervalos regulares. Para modelar la estacionalidad, se puede utilizar el modelo SARIMA:
  $
  (p, d, q) \times (P, D, Q)_s
  $

  donde $s$ es la periodicidad de la estacionalidad.

- **Ciclos**: Fluctuaciones que ocurren a intervalos irregulares debido a factores económicos o sociales.

- **Ruido**: Variaciones aleatorias que no pueden ser atribuidas a los componentes anteriores.

#### 2. Anomalías en Series Temporales

Las anomalías son puntos de datos que se desvían significativamente de la tendencia general de la serie. Pueden ser clasificadas como:

- **Anomalías puntuales**: Eventos aislados que se desvían del comportamiento esperado en un solo punto de tiempo.
  
- **Anomalías contextuales**: Variaciones que son anómalas en un contexto específico, como el aumento de la temperatura en invierno.

- **Anomalías de cambio de concepto**: Cambios en el patrón de comportamiento que afectan toda la serie.

Identificar estas anomalías es crucial para la detección de fraudes, el mantenimiento preventivo en sistemas industriales y la identificación de problemas de salud pública.

#### 3. Métodos de Detección de Anomalías

##### 3.1 Métodos Estadísticos

Los métodos estadísticos son enfoques tradicionales para la detección de anomalías. Algunas técnicas comunes incluyen:

- **Desviación Estándar**: Utilizando la media $\mu$ y la desviación estándar $\sigma$ para identificar puntos atípicos. Un valor $x_t$ se considera una anomalía si:

  $
  |x_t - \mu| > k \cdot \sigma
  $

  donde $k$ es típicamente 2 o 3.

- **Boxplots**: Los boxplots permiten visualizar la distribución de los datos. Un punto se considera anómalo si se encuentra fuera de los "bigotes":

  $
  \text{Límite superior} = Q3 + 1.5 \cdot IQR
  $
  
  $
  \text{Límite inferior} = Q1 - 1.5 \cdot IQR
  $

  donde $IQR = Q3 - Q1$ es el rango intercuartílico.

##### 3.2 Modelos Basados en Aprendizaje Automático

Los algoritmos de aprendizaje automático han revolucionado la detección de anomalías. Algunos métodos incluyen:

- **Isolation Forest**: Este modelo crea múltiples árboles de decisión y calcula la profundidad de cada punto. Los puntos que son fáciles de aislar se consideran anomalías. La puntuación de anomalía se calcula como:

  $
  \text{Anomaly Score} = \frac{2^H}{c(n)}
  $

  donde $H$ es la altura promedio del árbol y $c(n)$ es una constante de normalización basada en el número total de observaciones.

- **Autoencoders**: Son redes neuronales diseñadas para aprender una representación comprimida de los datos. Si la reconstrucción de un dato tiene un error significativo, se considera anómalo. La función de pérdida comúnmente utilizada es:

  $
  \text{Loss} = \frac{1}{N} \sum_{i=1}^{N} (x_i - \hat{x}_i)^2
  $

  donde $x_i$ es la entrada original y $\hat{x}_i$ es la salida reconstruida.

#### 4. Visualización de Anomalías

La visualización es una herramienta crucial en la detección de anomalías, ya que permite observar patrones de manera intuitiva:

- **Gráficos de Líneas**: Muestran la serie temporal y facilitan la identificación visual de anomalías.

- **Histogramas**: Muestran la distribución de los datos y ayudan a identificar outliers.

- **Diagramas de Cajas**: Resaltan visualmente los puntos fuera de los límites esperados, facilitando la detección de anomalías.

#### 5. Implementación de la Detección de Anomalías

En este proyecto, se implementa un sistema de detección de anomalías utilizando datos históricos de precios de Bitcoin. Los pasos clave incluyen:

1. **Cargar y Preprocesar Datos**: Obtención de datos a través de una API y limpieza de valores faltantes. Es esencial manejar los valores faltantes, que pueden afectar el análisis.

2. **Modelado**: Utilización de Isolation Forest para detectar anomalías. Este modelo aprende patrones a partir de los datos y asigna puntuaciones a cada punto en función de su capacidad para ser aislado.

3. **Evaluación**: Aunque puede ser difícil evaluar la precisión de la detección de anomalías sin etiquetas, se pueden utilizar métricas como el Error Absoluto Medio (MAE) en comparación con anomalías conocidas:

   $
   \text{MAE} = \frac{1}{N} \sum_{i=1}^{N} |y_i - \hat{y}_i|
   $

   donde $y_i$ es el valor real y $\hat{y}_i$ es el valor predicho.

### Ecuaciones Matemáticas

   $

   $

2. **Cálculo de la Desviación Estándar**:
   $
   \sigma = \sqrt{\frac{1}{N-1} \sum_{t=1}^{N} (x_t - \mu)^2}

   \text{Anomaly Score} = \frac{2^H}{c(n)}

   |x_t - \mu| > k \cdot \sigma

   IQR = Q3 - Q1
   $

   $

**Importar Bibliotecas Necesarias**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from sklearn.ensemble import IsolationForest
from sklearn.metrics import mean_absolute_error
```

**Cargar los Datos**
Usaremos yfinance para obtener datos históricos de precios de Bitcoin.

```python
# Descargar datos de Bitcoin
data = yf.download("BTC-USD", start="2020-01-01", end="2023-09-01", interval="1d")
data = data[['Close']]  # Usamos solo el precio de cierre
data.reset_index(inplace=True)
data.rename(columns={'Date': 'date', 'Close': 'price'}, inplace=True)
```

```python
# Visualizar los datos
plt.figure(figsize=(14, 7))
plt.plot(data['date'], data['price'], label='Precio de Bitcoin')
plt.title('Precio de Bitcoin')
plt.xlabel('Fecha')
plt.ylabel('Precio en USD')
plt.legend()
plt.show()
```

**Preprocesar los Datos**

Para la detección de anomalías, debemos asegurarnos de que los datos estén limpios y sin valores faltantes.

```python
# Rellenar valores faltantes (si los hay)
data['price'].fillna(method='ffill', inplace=True)

# Verificar si hay valores nulos
print(data.isnull().sum())
```

**Detección de Anomalías**
Usaremos el modelo Isolation Forest, que es un método popular para la detección de anomalías.

```python
# Preparar los datos para el modelo
X = data[['price']]

# Crear el modelo Isolation Forest
model = IsolationForest(contamination=0.05, random_state=42)
data['anomaly'] = model.fit_predict(X)

# Convertir -1 (anomalía) a 1 y 1 (normal) a 0
data['anomaly'] = data['anomaly'].map({1: 0, -1: 1})
```

```python
# Visualizar las anomalías
plt.figure(figsize=(14, 7))
plt.plot(data['date'], data['price'], label='Precio de Bitcoin')
plt.scatter(data[data['anomaly'] == 1]['date'], data[data['anomaly'] == 1]['price'], color='red', label='Anomalías', s=50)
plt.title('Detección de Anomalías en el Precio de Bitcoin')
plt.xlabel('Fecha')
plt.ylabel('Precio en USD')
plt.legend()
plt.show()
```

**Evaluación del Modelo**

Para evaluar el modelo, podemos calcular el error absoluto medio (MAE) entre las anomalías detectadas y los valores reales, aunque en este contexto, no hay etiquetas verdaderas para las anomalías.

```python
# Supongamos que tenemos una lista de anomalías conocidas para la evaluación (esto es solo un ejemplo)
known_anomalies = data[(data['date'] >= '2021-05-01') & (data['date'] <= '2021-05-10')]

# Calcular el MAE (esto es solo un ejemplo de cómo podríamos evaluar)
mae = mean_absolute_error(known_anomalies['price'], [known_anomalies['price'].mean()] * len(known_anomalies))
print(f'Mean Absolute Error: {mae}')
```

**Visualización de Resultados**

Finalmente, podemos visualizar el rendimiento del modelo de manera más detallada, incluyendo las anomalías detectadas en relación con el tiempo.

```python
plt.figure(figsize=(14, 7))
plt.plot(data['date'], data['price'], label='Precio de Bitcoin')
plt.scatter(data[data['anomaly'] == 1]['date'], data[data['anomaly'] == 1]['price'], color='red', label='Anomalías', s=50)
plt.axvline(x=pd.to_datetime('2021-05-01'), color='orange', linestyle='--', label='Anomalía Conocida')
plt.axvline(x=pd.to_datetime('2021-05-10'), color='orange', linestyle='--')
plt.title('Detección de Anomalías en el Precio de Bitcoin')
plt.xlabel('Fecha')
plt.ylabel('Precio en USD')
plt.legend()
plt.show()
```
