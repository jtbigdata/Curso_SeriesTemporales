<H1>Módulo 1: Introducción a las Series Temporales</H1>

**Conceptos clave**
Definición y características de una serie temporal:

**Serie Temporal:** Un conjunto de datos recogidos en intervalos regulares de tiempo. Ejemplos incluyen datos diarios de temperatura, ventas mensuales, o valores bursátiles.

**Características:** Los datos tienen una estructura temporal que puede ser analizada para identificar patrones y realizar previsiones.
Componentes de una serie temporal:

**Tendencia:** La dirección general a largo plazo de los datos. Puede ser ascendente, descendente, o constante.
Estacionalidad: Patrones recurrentes en intervalos regulares, como estacionalidades mensuales o anuales.
Ruido: Variaciones aleatorias y no sistemáticas en los datos que no pueden ser modeladas fácilmente.
Tipos de series temporales:

**Estacionarias:** Sus estadísticas (media, varianza) no cambian con el tiempo. Ejemplos incluyen series con estacionalidad estable y sin tendencia a largo plazo.

**No Estacionarias:** Sus estadísticas cambian con el tiempo. Pueden tener tendencias, estacionalidades cambiantes, o ambas.

# **Marco Teórico: Análisis Exploratorio de Series Temporales con Ecuaciones Matemáticas**

## **1. Definición y Características de una Serie Temporal**

Una **serie temporal**  $\{y_t\}$ es una secuencia de observaciones $y_t$ indexadas por el tiempo $t$. Los datos se organizan cronológicamente, permitiendo observar patrones y tendencias a lo largo del tiempo.

**Características Clave**:
- **Temporalidad**: Los datos se recogen en intervalos regulares de tiempo.
- **Dependencia Temporal**: Los valores actuales $y_t$ suelen estar correlacionados con valores pasados $y_{t-1}$, $y_{t-2}$, etc.

## **2. Componentes de una Serie Temporal**

Una serie temporal se puede descomponer en varios componentes principales:

1. **Tendencia $T_t$**:
   - **Definición**: La tendencia es la dirección general en la que los datos se mueven a largo plazo. Se modela como una función de tiempo $ t $.
   - **Ecuación**: Una tendencia lineal se puede representar como:
     $
     T_t = \beta_0 + \beta_1 t
     $
     donde $\beta_0$ es el intercepto y $\beta_1$ es la pendiente de la tendencia.

2. **Estacionalidad $S_t$**:
   - **Definición**: La estacionalidad representa patrones que se repiten en intervalos regulares, como estaciones del año o días de la semana.
   - **Ecuación**: Para una estacionalidad anual con período $p$, se puede modelar como:
     
     $
     S_t = \sum_{k=1}^{p} \left( \gamma_k \sin\left(\frac{2 \pi k t}{p}\right) + \delta_k \cos\left(\frac{2 \pi k t}{p}\right) \right)
     $
     
     donde $ \gamma_k$ y $\delta_k$ son coeficientes que capturan la magnitud de los patrones estacionales.

3. **Ruido $\epsilon_t $**:
   - **Definición**: El ruido es la variabilidad aleatoria que no puede ser explicada por la tendencia y la estacionalidad. Representa fluctuaciones imprevistas en los datos.
   - **Características**:
     - **Aleatoriedad**: El ruido es impredecible y no sigue un patrón específico.
     - **Media Cero**: En muchos casos, se asume que el ruido tiene una media de cero.
     - **Varianza Constante**: Puede tener una varianza constante o cambiar en función del tiempo.
   - **Ecuación**: El ruido se modela comúnmente como una variable aleatoria con media cero y varianza constante:
     
     $
     \epsilon_t \sim \text{N}(0, \sigma^2)
     $
     donde $\sigma^2$ es la varianza del ruido.

## **3. Tipos de Series Temporales**

- **Estacionarias**:
  - **Definición**: Una serie temporal es estacionaria si sus propiedades estadísticas, como la media y la varianza, permanecen constantes a lo largo del tiempo.
  - **Ecuación**: Para una serie estacionaria $y_t$, la forma general es:
    $
    y_t = T_t + S_t + \epsilon_t
    $
    y las propiedades estadísticas no dependen de $t$.

- **No Estacionarias**:
  - **Definición**: Una serie no estacionaria muestra variaciones en la media y/o la varianza con el tiempo. Pueden presentar tendencias o estacionalidades cambiantes.
  - **Ecuación**: Para una serie con tendencia, la serie temporal no es estacionaria y se puede expresar como:
    $
    y_t = T_t + S_t + \epsilon_t
    $
    donde $T_t$ varía con $t$.

## **4. Análisis Exploratorio de Series Temporales**

1. **Visualización de Datos**:
   - **Objetivo**: Graficar $y_t$ para observar patrones, tendencias y componentes.

2. **Descomposición de Series Temporales**:
   - **Método de Descomposición**:
     La serie temporal $y_t$ puede descomponerse en sus componentes básicos usando un modelo aditivo o multiplicativo:
     
     **Modelo Aditivo**:
     $
     y_t = T_t + S_t + \epsilon_t
     $
     
     **Modelo Multiplicativo**:
     $
     y_t = T_t \times S_t \times \epsilon_t
     $
     
     - **Descomposición Aditiva**:
       La serie se descompone en componentes aditivos, y los residuos se calculan como:
       $
       \text{Residuos} = y_t - (T_t + S_t)
       $
     
     - **Descomposición Multiplicativa**:
       La serie se descompone en componentes multiplicativos. En este caso, los residuos se calculan como:
       $
       \text{Residuos} = \frac{y_t}{T_t \times S_t}
       $

3. **Interpretación de Resultados**:
   - **Tendencia**: Evaluar la dirección general y el comportamiento a largo plazo de la serie.
   - **Estacionalidad**: Identificar patrones periódicos y ajustar modelos predictivos en consecuencia.
   - **Ruido**: Analizar la variabilidad que no se puede atribuir a la tendencia o estacionalidad. El análisis del ruido puede revelar la calidad de los datos y la precisión de los modelos.

## **Conclusión**

El análisis exploratorio de series temporales es esencial para entender la estructura y los patrones en datos temporales. Descomponer la serie en componentes básicos (tendencia, estacionalidad y ruido) y visualizarlos proporciona una comprensión más profunda y permite mejorar la modelización y previsión. Las ecuaciones matemáticas ayudan a cuantificar y analizar estos componentes, facilitando una interpretación más rigurosa de los datos temporales.

<H1>Proyecto: Análisis Exploratorio de una Serie Temporal </H1>

Vamos a realizar un análisis exploratorio de una serie temporal descargando un conjunto de datos de ventas mensuales. Utilizaremos Python y pandas para el análisis, y matplotlib y statsmodels para la visualización.

**Instalar y cargar las librerías necesarias:** Primero, importaremos las librerías necesarias. En un entorno donde no podemos instalar librerías directamente, podemos cargarlas desde la web.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.seasonal import seasonal_decompose
```

**Descargar el conjunto de datos:** Para este proyecto, usaremos un conjunto de datos de ventas mensuales. Podemos utilizar datos disponibles en línea.

```python
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
data = pd.read_csv(url, header=0, index_col=0, parse_dates=True, squeeze=True)
```

**Paso 2: Análisis Descriptivo Inicial**

**Visualizar los datos:** Graficamos los datos para entender su estructura.

```python
plt.figure(figsize=(12, 6))
plt.plot(data, label='Número de pasajeros')
plt.title('Número de Pasajeros Aéreos Mensuales')
plt.xlabel('Fecha')
plt.ylabel('Número de Pasajeros')
plt.legend()
plt.show()
```

**Descomponer la serie temporal:** Utilizamos la descomposición estacional para identificar la tendencia, estacionalidad y ruido.

```python
decomposition = seasonal_decompose(data, model='multiplicative')

plt.figure(figsize=(12, 12))

plt.subplot(4, 1, 1)
plt.plot(data, label='Original')
plt.legend(loc='upper left')

plt.subplot(4, 1, 2)
plt.plot(decomposition.trend, label='Tendencia')
plt.legend(loc='upper left')

plt.subplot(4, 1, 3)
plt.plot(decomposition.seasonal, label='Estacionalidad')
plt.legend(loc='upper left')

plt.subplot(4, 1, 4)
plt.plot(decomposition.resid, label='Residuos')
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

**Analizar los componentes:**

**Tendencia:** Observa si hay una tendencia clara en el tiempo.

**Estacionalidad:** Identifica patrones repetitivos en los datos.

**Ruido:** Examina la variabilidad que no puede ser explicada por la tendencia y la estacionalidad.
