<H1>Marco Teórico</H1>

#### 1. Introducción a las Series Temporales

Las series temporales son secuencias de datos recogidos o registrados en momentos sucesivos, donde el tiempo es una variable crucial que puede influir en el comportamiento de los datos. Estos datos pueden ser observados en diversas áreas como la economía, la finanza, la meteorología y las ciencias sociales. El análisis de series temporales se centra en identificar patrones a lo largo del tiempo, comprender las dinámicas subyacentes y predecir futuros comportamientos basados en valores pasados.

#### 2. Modelos Multivariados

##### 2.1. VAR (Vector Autoregression)

El modelo VAR es una técnica fundamental para el análisis de series temporales multivariadas. Permite modelar múltiples variables interdependientes, capturando tanto la dinámica de cada serie como sus interrelaciones. A diferencia de los modelos univariantes, el VAR considera que cada variable se ve afectada no solo por sus propios rezagos, sino también por los rezagos de otras variables.

La forma general de un modelo VAR de orden $p$ se expresa como:

$
Y_t = A_1 Y_{t-1} + A_2 Y_{t-2} + \ldots + A_p Y_{t-p} + \epsilon_t
$

donde:
- $Y_t$ es un vector de variables endógenas en el tiempo $t$, representando múltiples series temporales.
- $A_i$ son matrices de coeficientes que representan la influencia de los rezagos de las variables sobre las variables actuales.
- $\epsilon_t$ es un vector de errores que se supone normalmente distribuido con media cero y una matriz de covarianza $\Sigma$.

El modelo VAR permite analizar la causalidad entre las variables a través de la estimación de los coeficientes, lo que permite inferir cómo cambios en una variable pueden afectar a otras en el sistema.

##### 2.2. VECM (Vector Error Correction Model)

Cuando las series son cointegradas, el modelo de corrección de errores (VECM) es más apropiado. Este modelo no solo captura las dinámicas de corto plazo entre las variables, sino que también incorpora una relación de equilibrio a largo plazo.

Un VECM se expresa como:

$
\Delta Y_t = \alpha (\beta' Y_{t-1}) + \sum_{i=1}^{p-1} \Gamma_i \Delta Y_{t-i} + \epsilon_t
$

donde:
- $\Delta Y_t $ representa el cambio en las variables.
- $\alpha$ es un vector que indica la velocidad de ajuste hacia el equilibrio a largo plazo.
- $\beta$ es un vector que representa la relación de cointegración.
- $\Gamma_i$  son matrices de coeficientes que modelan las relaciones dinámicas entre las diferencias de las variables.

La inclusión del término de corrección de errores permite que el modelo ajuste las predicciones a largo plazo basándose en los errores de predicción previos.

#### 3. Cointegración

La cointegración es un concepto clave en el análisis de series temporales, que implica que, aunque las series individuales puedan ser no estacionarias (es decir, sus propiedades estadísticas cambian a lo largo del tiempo), una combinación lineal de ellas puede ser estacionaria.

##### 3.1. Prueba de Cointegración

Una de las pruebas más comunes para determinar la cointegración es la prueba de Engle-Granger. Este método implica los siguientes pasos:

1. **Regresión Inicial**: Se realiza una regresión de una serie sobre otra, por ejemplo:

   $
   Y_t = \alpha + \beta X_t + u_t
   $

2. **Verificación de Estacionariedad**: Se aplica una prueba, como la prueba de Dickey-Fuller aumentada, a los residuos $u_t$ para verificar si son estacionarios. Si los residuos son estacionarios, se concluye que $Y_t $ y $ X_t $ están cointegrados.

Las series cointegradas sugieren una relación de equilibrio, permitiendo modelar sus interacciones de forma más efectiva mediante VECM.

##### 3.2. Propiedades de la Cointegración

- **Relación a Largo Plazo**: Las series cointegradas pueden divergir en el corto plazo, pero eventualmente tienden a volver a su relación de equilibrio a largo plazo.
- **Estimación de Parámetros**: La cointegración permite una estimación más robusta de los parámetros en modelos de regresión, ya que las relaciones a largo plazo son más relevantes en contextos económicos.

#### 4. Preprocesamiento de Datos

El preprocesamiento es esencial en el análisis de series temporales. Incluye pasos que garantizan que los datos sean adecuados para el modelado. Esto puede incluir:

- **Rellenar valores faltantes**: Utilizando métodos como la imputación por la media, la interpolación o el uso de modelos predictivos.
- **Transformaciones**: Aplicar transformaciones logarítmicas o diferenciación para estabilizar la varianza y hacer las series más adecuadas para el modelado.

La transformación logarítmica de una serie temporal $Y$ se define como:

$
Z_t = \log(Y_t)
$

Esto puede ayudar a reducir la heterocedasticidad y facilitar la normalización de los datos.

#### 5. Análisis de Autocorrelación

Los gráficos de autocorrelación (ACF) y autocorrelación parcial (PACF) son herramientas útiles para identificar la estructura de rezago en series temporales. La ACF mide la correlación entre una serie y sus rezagos, mientras que la PACF mide la correlación de un rezago específico mientras se controlan los efectos de los rezagos intermedios.

La ACF se define como:

$
\rho_k = \frac{\text{Cov}(Y_t, Y_{t-k})}{\sigma_Y^2}
$

donde:
- $\rho_k$ es la autocorrelación en el rezago $k$.
- $\text{Cov}$ es la covarianza entre las observaciones.
- $\sigma_Y^2$  es la varianza de la serie.

La PACF se define de manera similar, pero ajusta por los efectos de rezagos intermedios, lo que proporciona una visión más clara de la relación directa entre la serie y sus rezagos.

#### 6. Validación de Modelos

La validación de modelos en series temporales es crucial para garantizar que las predicciones sean precisas y confiables. Algunas métricas comunes incluyen:

- **Error Cuadrático Medio (MSE)**:
  $
  MSE = \frac{1}{n} \sum_{t=1}^n (Y_t - \hat{Y}_t)^2
  $
  
- **Error Absoluto Medio (MAE)**:
  $
  MAE = \frac{1}{n} \sum_{t=1}^n |Y_t - \hat{Y}_t|
  $

Donde $Y_t$ es el valor real y $\hat{Y}_t$ es el valor predicho.

#### 7. Conclusión

Este marco teórico proporciona una base sólida para el análisis de series temporales multivariadas. La implementación de modelos VAR y VECM, la verificación de cointegración y el adecuado preprocesamiento de datos son pasos cruciales para entender las relaciones dinámicas entre variables económicas y realizar predicciones informadas. Las técnicas de autocorrelación ofrecen información adicional sobre la estructura temporal de las series, lo que es esencial para la modelización efectiva. A través de un análisis riguroso y métodos estadísticos apropiados, es posible descubrir patrones significativos y relaciones que pueden informar decisiones estratégicas en diversas disciplinas.

**Importar las librerías necesarias**

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.api import VAR
from sklearn.preprocessing import FunctionTransformer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
```

**Descargar datos**

Vamos a descargar un conjunto de datos de series temporales. En este caso, usaremos datos de precios de acciones de diferentes compañías.

```python
# Definir las acciones a descargar
tickers = ['AAPL', 'MSFT', 'GOOGL']

# Descargar datos
data = yf.download(tickers, start='2020-01-01', end='2023-01-01')['Adj Close']
data = data.dropna()  # Eliminar filas con datos faltantes
```

```python
data.head()
```

**Preprocesamiento de datos**
Vamos a implementar un pipeline para preprocesar los datos.

```python
# Definir el transformador
class LogTransform(FunctionTransformer):
    def __init__(self):
        super().__init__(func=lambda x: np.log(x), validate=False)

# Crear un pipeline
preprocessor = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),  # Rellenar NaN con la media
    ('log_transform', LogTransform())  # Transformación logarítmica
])

# Aplicar el pipeline
data_transformed = preprocessor.fit_transform(data)
data_transformed = pd.DataFrame(data_transformed, columns=data.columns, index=data.index)
```

```python
print(data_transformed.head())
```

**Verificar estacionariedad**

```python
def test_stationarity(timeseries):
    # Determinar la media y la desviación estándar
    rolmean = timeseries.rolling(window=12).mean()
    rolstd = timeseries.rolling(window=12).std()

    # Graficar la serie temporal
    plt.figure(figsize=(12, 6))
    plt.plot(timeseries, color='blue', label='Original')
    plt.plot(rolmean, color='red', label='Media Móvil')
    plt.plot(rolstd, color='black', label='Desviación Estándar Móvil')
    plt.legend(loc='best')
    plt.title('Media y Desviación Estándar Móvil')
    plt.show()
```

```python
# Probar estacionariedad en las series transformadas
for column in data_transformed.columns:
    test_stationarity(data_transformed[column])
```

**Ajustar el modelo VAR**

```python
# Ajustar el modelo VAR
model = VAR(data_transformed)
results = model.fit(maxlags=15, ic='aic')
```

```python
# Resumen de los resultados
print(results.summary())
```

**Análisis de los resultados**

```python
# Graficar la ACF y PACF para cada serie temporal
for column in data_transformed.columns:
    plt.figure(figsize=(12, 6))
    sm.graphics.tsa.plot_acf(data_transformed[column], lags=40)
    plt.title(f'ACF para {column}')
    plt.show()

    plt.figure(figsize=(12, 6))
    sm.graphics.tsa.plot_pacf(data_transformed[column], lags=40)
    plt.title(f'PACF para {column}')
    plt.show()
```

**Cointegración**
