**Importar Librerías**
Cargamos las librerías requeridas para la implementación de los modelos ARIMA, Prophet, XGBoost, y para la optimización de hiperparámetros.

```python
import yfinance as yf
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.model_selection import train_test_split, GridSearchCV
from xgboost import XGBRegressor
from prophet import Prophet
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt
```

**Descargar el Dataset desde Yahoo Finance**
Descargamos los datos de una acción, por ejemplo, de Apple (AAPL), para un rango de fechas determinado.

```python
# Descargar datos de la acción de Apple desde 2010
data = yf.download('AAPL', start='2010-01-01', end='2023-01-01')
data = data[['Adj Close']]
data.rename(columns={'Adj Close': 'price'}, inplace=True)
data.head()
```

**Preparación del Dataset**
Convertimos el dataset para su uso tanto en el modelo ARIMA como en Prophet y XGBoost.

```python
# Preparar los datos para Prophet
data_prophet = data.reset_index().rename(columns={'Date': 'ds', 'price': 'y'})

# Preparar los datos para ARIMA + XGBoost
data['lag1'] = data['price'].shift(1)
data.dropna(inplace=True)

# Crear características adicionales para XGBoost
X = data.drop(['price'], axis=1)
y = data['price']

# Dividir en datos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

**Implementación del Modelo Prophet**
Prophet es un modelo desarrollado por Facebook que se utiliza para la predicción de series temporales con componentes estacionales y tendencias.

```python
# Definir y ajustar el modelo Prophet
model_prophet = Prophet()
model_prophet.fit(data_prophet)

# Realizar predicciones
future = model_prophet.make_future_dataframe(periods=365)
forecast = model_prophet.predict(future)

# Graficar los resultados
fig = model_prophet.plot(forecast)
```

**Implementación del Modelo ARIMA**
ARIMA es un modelo popular para series temporales que capta tanto las tendencias como las diferencias estacionales de los datos.

```python
# Definir y ajustar el modelo ARIMA
model_arima = ARIMA(y_train, order=(5, 1, 0))
arima_fit = model_arima.fit()

# Realizar predicciones
arima_pred = arima_fit.forecast(steps=len(y_test))

# Evaluar el modelo ARIMA
mse_arima = mean_squared_error(y_test, arima_pred)
print(f'MSE ARIMA: {mse_arima}')
```

**Implementación del Modelo Híbrido ARIMA + XGBoost**
Aquí combinamos las predicciones del modelo ARIMA con el modelo XGBoost para mejorar las predicciones.

```python
# Entrenar el modelo XGBoost
xgb_model = XGBRegressor(n_estimators=100, learning_rate=0.1)
xgb_model.fit(X_train, y_train)

# Predecir con el modelo XGBoost
xgb_pred = xgb_model.predict(X_test)

# Combinar las predicciones de ARIMA y XGBoost
hybrid_pred = (xgb_pred + arima_pred) / 2

# Evaluar el modelo híbrido
mse_hybrid = mean_squared_error(y_test, hybrid_pred)
print(f'MSE Modelo Híbrido: {mse_hybrid}')
```

**Optimización de Hiperparámetros**
Podemos optimizar los hiperparámetros del modelo XGBoost utilizando GridSearchCV.

```python
# Definir el espacio de búsqueda de hiperparámetros
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2]
}

# Realizar la búsqueda de hiperparámetros
grid_search = GridSearchCV(XGBRegressor(), param_grid, cv=3, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

# Mejor modelo
best_xgb = grid_search.best_estimator_
best_xgb_pred = best_xgb.predict(X_test)

# Evaluar el mejor modelo
mse_best_xgb = mean_squared_error(y_test, best_xgb_pred)
print(f'MSE XGBoost Optimizado: {mse_best_xgb}')
```

**Comparación de Modelos** Finalmente, comparamos las métricas de los diferentes modelos.

```python
# Métricas de evaluación
mse_prophet = mean_squared_error(data_prophet['y'].iloc[-len(y_test):], forecast['yhat'].iloc[-len(y_test):])

print(f'MSE Prophet: {mse_prophet}')
print(f'MSE ARIMA: {mse_arima}')
print(f'MSE XGBoost: {mse_best_xgb}')
print(f'MSE Modelo Híbrido: {mse_hybrid}')
```

**Visualización de Resultados** Para visualizar las predicciones de los modelos y la comparación con los datos reales.

```python
plt.figure(figsize=(10,6))
plt.plot(y_test.index, y_test, label='Real')
plt.plot(y_test.index, arima_pred, label='ARIMA')
plt.plot(y_test.index, xgb_pred, label='XGBoost')
plt.plot(y_test.index, hybrid_pred, label='Híbrido ARIMA + XGBoost')
plt.plot(data_prophet['ds'].iloc[-len(y_test):], forecast['yhat'].iloc[-len(y_test):], label='Prophet')
plt.legend()
plt.title('Comparación de Modelos')
plt.show()
```
