<H1>Marco Teórico: Modelos Estacionales SARIMA y SARIMAX</H1>

Los modelos ARIMA y sus extensiones estacionales, SARIMA y SARIMAX, son ampliamente utilizados para predecir series temporales que presentan características como tendencia, estacionalidad y, en algunos casos, relaciones con variables exógenas. En este marco teórico, se explicarán los fundamentos de estos modelos, su estructura matemática y su aplicación práctica.

**1. Modelos ARIMA (AutoRegressive Integrated Moving Average)**

El modelo **ARIMA** es un enfoque clásico para modelar series temporales que tienen componentes no estacionarios. Se descompone en tres partes:
- **AR (AutoRegresión)**: Describe la relación lineal entre los valores pasados y actuales.
- **I (Diferenciación)**: Se usa para transformar una serie no estacionaria en estacionaria.
- **MA (Media Móvil)**: Modela el error en función de los errores anteriores.

El modelo **ARIMA** está parametrizado por tres valores enteros $(p, d, q)$:

- **$p$**: Orden del modelo autoregresivo (AR), que refleja el número de retardos (lags) de la variable dependiente.
- **$d$**: Número de diferencias necesarias para hacer la serie estacionaria.
- **$q$**: Orden del modelo de media móvil (MA), que refleja el número de errores pasados.

La ecuación general de un modelo ARIMA es:

$
y_t = c + \sum_{i=1}^{p} \phi_i y_{t-i} + \sum_{j=1}^{q} \theta_j \epsilon_{t-j} + \epsilon_t
$

Donde:
- $y_t$ es el valor en el tiempo $t$,
- $\phi_i$ son los coeficientes del término autoregresivo (AR),
- $\theta_j$ son los coeficientes del término de media móvil (MA),
- $\epsilon_t$ es el error en \(t\),
- $c$ es una constante.

**2. Estacionalidad y el Modelo SARIMA (Seasonal ARIMA)**

Cuando una serie temporal tiene un comportamiento repetitivo en intervalos regulares (por ejemplo, ventas anuales con estacionalidad mensual), se requiere un modelo que pueda capturar este patrón estacional. El modelo **SARIMA** extiende el ARIMA al agregar un componente estacional.

El modelo SARIMA se parametriza como $(p,d,q)(P,D,Q,S)$, donde:
- **$P, D, Q$**: Son los parámetros del modelo estacional, equivalentes a \(p, d, q\), pero aplicados al componente estacional.
- **$S$**: Es el período de la estacionalidad (por ejemplo, 12 para datos mensuales con estacionalidad anual).

La ecuación de un modelo **SARIMA** es:

$
\Phi_P(B^S)(1 - B^S)^D y_t = \theta_q(B) \epsilon_t
$

Donde:
- $B$ es el operador de retardos (backshift),
- $\Phi_P(B^S)$ es el componente autoregresivo estacional,
- $(1 - B^S)^D$ es la diferencia estacional,
- $\theta_q(B)$ es el componente de media móvil no estacional.

**3. Diferenciación y Estacionariedad**

El concepto de **estacionariedad** es clave en la modelización de series temporales. Una serie es estacionaria si sus propiedades estadísticas, como la media y la varianza, son constantes en el tiempo. Para verificar si una serie es estacionaria, utilizamos la **prueba de Dickey-Fuller Aumentada (ADF)**, que tiene la siguiente hipótesis:
- **Hipótesis nula** (\(H_0\)): La serie no es estacionaria.
- **Hipótesis alternativa** (\(H_1\)): La serie es estacionaria.

Si el valor $p$ de la prueba ADF es mayor que 0.05, no podemos rechazar $H_0$, y la serie es considerada no estacionaria.

Para hacer la serie estacionaria, aplicamos **diferenciación**:

$
y'_t = y_t - y_{t-1}
$

Si se requiere más de una diferenciación, se puede aplicar varias veces.

**4. Modelo SARIMAX (Seasonal ARIMA with Exogenous Variables)**

El modelo **SARIMAX** es una extensión del SARIMA que permite la inclusión de **variables exógenas** (independientes) que puedan influir en la serie temporal. Estas variables exógenas, denotadas como \(X_t\), proporcionan información adicional que mejora las predicciones.

La ecuación general del modelo SARIMAX es:

$
y_t = c + \sum_{i=1}^{p} \phi_i y_{t-i} + \sum_{j=1}^{q} \theta_j \epsilon_{t-j} + \sum_{k=1}^{r} \beta_k X_{t-k} + \epsilon_t
$

Donde:
- $X_{t-k}$ son las variables exógenas retardadas,
- $\beta_k$ son los coeficientes de las variables exógenas.

**5. Evaluación del Modelo: RMSE (Root Mean Squared Error)**

El **RMSE** es una métrica utilizada para evaluar la precisión de las predicciones de un modelo. El **RMSE** se calcula como la raíz cuadrada de la media de los errores cuadrados entre los valores reales y los predichos:

$
\text{RMSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y_i})^2}
$

Donde:
- $y_i$ son los valores observados,
- $\hat{y_i}$ son los valores predichos,
- $n$ es el número total de observaciones.

Un RMSE más bajo indica un mejor ajuste del modelo a los datos.

#### **6. Implementación Práctica: SARIMA y SARIMAX**

En la implementación práctica, dividimos los datos en un conjunto de entrenamiento y uno de prueba, ajustamos el modelo SARIMA o SARIMAX a los datos de entrenamiento, y luego generamos predicciones para el conjunto de prueba. Posteriormente, calculamos el **RMSE** para evaluar el rendimiento del modelo.

#### **Conclusión**

El modelo **SARIMA** es ideal para series temporales con estacionalidad, mientras que el **SARIMAX** agrega flexibilidad al incorporar variables exógenas. Ambos modelos dependen de que la serie sea estacionaria, lo que se puede verificar y corregir mediante la prueba ADF y la diferenciación. La métrica **RMSE** permite cuantificar el error de las predicciones y ajustar el modelo para obtener mejores resultados.

<H1>Proyecto</H1>

**Paso 1: Importar las Librerías Necesarias**

Primero, importaremos todas las librerías que usaremos para manipular los datos, visualizar y crear los modelos SARIMA y SARIMAX.

```python
# Importamos las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Para modelos ARIMA y SARIMAX
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.arima.model import ARIMA

# Para pruebas de estacionariedad y descomposición de la serie temporal
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose

# Para dividir los datos y medir el rendimiento
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")
```

**Paso 2: Cargar y Visualizar los Datos**

Para este ejemplo, utilizaremos un dataset de ventas mensuales de una tienda, que puedes descargar desde aquí (ventas de shampoo mensuales).

```python
# Cargar el dataset de ventas mensuales
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/shampoo.csv'
df = pd.read_csv(url, parse_dates=['Month'], index_col='Month')
```

```python
# Renombrar la columna para facilitar el acceso
df.columns = ['Ventas']

# Graficar la serie temporal
plt.figure(figsize=(10,6))
plt.plot(df['Ventas'], color='blue')
plt.title('Ventas Mensuales de Shampoo')
plt.xlabel('Fecha')
plt.ylabel('Ventas')
plt.grid(True)
plt.show()
```

**Paso 3: Comprobar Estacionalidad y Estacionariedad**

Usamos la descomposición de la serie temporal para identificar si la serie tiene estacionalidad y tendencia.

```python
# Descomponer la serie en componentes (tendencia, estacionalidad y ruido)
decomposition = seasonal_decompose(df['Ventas'], model='multiplicative', period=12)

# Graficar los componentes
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10,8))

decomposition.trend.plot(ax=ax1, color='blue')
ax1.set_title('Tendencia')

decomposition.seasonal.plot(ax=ax2, color='green')
ax2.set_title('Estacionalidad')

decomposition.resid.plot(ax=ax3, color='red')
ax3.set_title('Residuo')

plt.tight_layout()
plt.show()
```

Para comprobar si la serie es estacionaria, aplicamos la Prueba de Dickey-Fuller.

```python
# Prueba de Dickey-Fuller
result = adfuller(df['Ventas'].dropna())
print('ADF Statistic:', result[0])
print('p-value:', result[1])

# Si el p-value es mayor a 0.05, la serie no es estacionaria
```

El resultado de la prueba de Dickey-Fuller sugiere que la serie temporal no es estacionaria. Esto se deduce porque el estadístico ADF es positivo (3.06) y el valor p es 1.0, lo que es mucho mayor que el umbral común de 0.05. En este caso, debemos aplicar técnicas de diferenciación para hacer la serie estacionaria.

**Paso 4: Diferenciar la Serie para Hacerla Estacionaria (si es necesario)**

Si la serie no es estacionaria, podemos aplicar diferencias para volverla estacionaria.

```python
# Aplicar diferencia para hacer estacionaria (si no lo es)
df['Ventas_diff'] = df['Ventas'].diff()

# Graficar la serie diferenciada
plt.figure(figsize=(10,6))
plt.plot(df['Ventas_diff'], color='orange')
plt.title('Ventas Mensuales (Diferenciadas)')
plt.show()
```

**Paso 5: Implementar SARIMA**

El modelo SARIMA extiende el ARIMA incorporando estacionalidad. Ahora ajustamos el modelo SARIMA.

```python
# Dividimos los datos en conjunto de entrenamiento y prueba
train, test = train_test_split(df['Ventas'], test_size=0.2, shuffle=False)

# Ajustamos un modelo SARIMA (p,d,q)(P,D,Q, S)
sarima_model = SARIMAX(train, order=(1,1,1), seasonal_order=(1,1,1,12))
sarima_fit = sarima_model.fit(disp=False)

# Predecimos en el conjunto de prueba
sarima_pred = sarima_fit.predict(start=len(train), end=len(train)+len(test)-1, dynamic=False)

# Graficamos los resultados
plt.figure(figsize=(10,6))
plt.plot(train.index, train, label='Entrenamiento')
plt.plot(test.index, test, label='Prueba')
plt.plot(test.index, sarima_pred, label='Predicciones SARIMA', color='red')
plt.legend()
plt.title('Predicciones usando SARIMA')
plt.show()

# Evaluar el modelo con RMSE
sarima_rmse = np.sqrt(mean_squared_error(test, sarima_pred))
print('SARIMA RMSE:', sarima_rmse)
```

**Paso 6: Implementar SARIMAX con Variables Exógenas**

Para SARIMAX, necesitamos una variable exógena. Aquí podemos simular una variable exógena, como una campaña publicitaria que influye en las ventas.

```python
# Crear una variable exógena (ejemplo: campaña publicitaria)
np.random.seed(42)
df['Campaña'] = np.random.normal(100, 10, len(df))

# Dividimos el conjunto de entrenamiento y prueba con la variable exógena
train_X = df[['Campaña']].iloc[:len(train)]
test_X = df[['Campaña']].iloc[len(train):]

# Ajustamos un modelo SARIMAX con variables exógenas
sarimax_model = SARIMAX(train, exog=train_X, order=(1,1,1), seasonal_order=(1,1,1,12))
sarimax_fit = sarimax_model.fit(disp=False)

# Predecimos en el conjunto de prueba
sarimax_pred = sarimax_fit.predict(start=len(train), end=len(train)+len(test)-1, exog=test_X)

# Graficamos los resultados
plt.figure(figsize=(10,6))
plt.plot(train.index, train, label='Entrenamiento')
plt.plot(test.index, test, label='Prueba')
plt.plot(test.index, sarimax_pred, label='Predicciones SARIMAX', color='green')
plt.legend()
plt.title('Predicciones usando SARIMAX con variables exógenas')
plt.show()

# Evaluar el modelo con RMSE
sarimax_rmse = np.sqrt(mean_squared_error(test, sarimax_pred))
print('SARIMAX RMSE:', sarimax_rmse)
```

**Paso 7: Comparación y Conclusión**

Finalmente, comparamos los resultados de los modelos SARIMA y SARIMAX.

```python
print(f'RMSE SARIMA: {sarima_rmse}')
print(f'RMSE SARIMAX: {sarimax_rmse}')

if sarimax_rmse < sarima_rmse:
    print("El modelo SARIMAX con variables exógenas mejora las predicciones.")
else:
    print("El modelo SARIMA es mejor para estas predicciones.")
```
