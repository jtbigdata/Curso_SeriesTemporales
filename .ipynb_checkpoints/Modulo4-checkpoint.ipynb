<H1>Marco Teórico: Modelos de Suavizamiento Exponencial y Holt-Winters</H1>


El análisis de series temporales es esencial para predecir y modelar datos dependientes del tiempo, como las ventas o la producción industrial. Los modelos de **suavizamiento exponencial** y **Holt-Winters** son herramientas estadísticas clave para capturar tendencias y estacionalidades en este tipo de datos. A continuación, se presenta un marco teórico con las ecuaciones en formato LaTeX para facilitar su integración.

---

### **1. Suavizamiento Exponencial Simple (SES)**

#### **Definición**

El **suavizamiento exponencial simple (SES)** es un método que genera predicciones basadas en una media ponderada de las observaciones pasadas, donde las más recientes reciben mayor peso. Es útil para series temporales sin tendencia ni estacionalidad.

#### **Ecuación de Suavizamiento Exponencial Simple**

El valor suavizado en el tiempo $ t $, denotado como $S_t$, se calcula como una combinación ponderada del valor observado $y_t$ en el tiempo $t$ y el valor suavizado anterior $S_{t-1}$:

$
S_t = \alpha y_t + (1 - \alpha) S_{t-1}
$

Donde:
- $ S_t $ es el valor suavizado en el tiempo $ t $.
- $ y_t $ es el valor observado en el tiempo $ t $.
- $ \alpha $ es el parámetro de suavizamiento, con $ 0 < \alpha \leq 1 $.
- $ S_{t-1} $ es el valor suavizado en el tiempo $ t-1 $.

La constante $ \alpha $ determina el peso que se otorga a las observaciones más recientes frente a las observaciones más antiguas. Un $ \alpha $ cercano a 1 otorga mayor importancia a los valores recientes.

---

### **2. Modelo Holt-Winters**

#### **Definición**

El modelo de **Holt-Winters** extiende el suavizamiento exponencial para incluir componentes de **tendencia** y **estacionalidad**. Es adecuado para series temporales con una estructura estacional repetitiva. Se presenta en dos variantes: **aditivo** (para estacionalidad constante) y **multiplicativo** (para estacionalidad proporcional al nivel de la serie).

#### **Componentes del Modelo Holt-Winters**

El modelo se basa en tres ecuaciones principales, que calculan:
1. **Nivel $ \ell_t $**: Representa el valor base de la serie.
2. **Tendencia $ b_t $**: Captura el cambio en el nivel de la serie a lo largo del tiempo.
3. **Estacionalidad $s_t$**: Representa los patrones estacionales recurrentes.

Las ecuaciones para actualizar estos componentes son:

$
\ell_t = \alpha \frac{y_t}{s_{t-m}} + (1 - \alpha)(\ell_{t-1} + b_{t-1})
$

$
b_t = \beta (\ell_t - \ell_{t-1}) + (1 - \beta) b_{t-1}
$

$
s_t = \gamma \frac{y_t}{\ell_t} + (1 - \gamma) s_{t-m}
$

Donde:
- $\ell_t$ es el nivel en el tiempo $t$.
- $b_t$ es la tendencia en el tiempo $t$.
- $s_t$ es el factor estacional en el tiempo $t$.
- $y_t$ es el valor observado en el tiempo $t$.
- $m$  es el periodo estacional.
- $ \alpha $, $ \beta $,  $\gamma$  son los parámetros de suavizamiento para nivel, tendencia y estacionalidad, respectivamente, con $ 0 < \alpha, \beta, \gamma \leq 1 $.

#### **Predicción con el Modelo Holt-Winters**

Para predecir $h$  periodos en el futuro, la fórmula es:

$
\hat{y}_{t+h} = (\ell_t + h b_t) s_{t+h-m(k+1)}
$

Donde:
- $ \hat{y}_{t+h}$ es la predicción para $h$  pasos adelante.
- $ k $ es el número de ciclos completos de estacionalidad en $h$.

---

### **3. Métricas de Evaluación**

#### **Error Cuadrático Medio (MSE)**

El **error cuadrático medio (MSE)** mide el promedio de los errores al cuadrado entre los valores predichos y los valores reales. Se expresa como:

$
MSE = \frac{1}{n} \sum_{t=1}^{n} (y_t - \hat{y}_t)^2
$

Donde:
- $ y_t$ es el valor real en el tiempo $t$.
- $\hat{y}_t$ es el valor predicho en el tiempo $t$.
- $n$ es el número de observaciones.

Un menor MSE indica un mejor ajuste del modelo.

#### **Error Absoluto Medio (MAE)**

El **error absoluto medio (MAE)** mide el promedio de las diferencias absolutas entre los valores reales y los predichos. Se expresa como:

$
MAE = \frac{1}{n} \sum_{t=1}^{n} |y_t - \hat{y}_t|
$

El MAE indica el error promedio en las mismas unidades que los datos originales.

---

### **4. Predicción a Futuro**

El modelo Holt-Winters permite realizar predicciones a futuro basadas en los componentes de nivel, tendencia y estacionalidad. Estas predicciones son útiles en aplicaciones como la planificación de la demanda, gestión de inventarios y predicción de ventas.

---

### **Conclusión**

El **suavizamiento exponencial simple** y el **modelo Holt-Winters** son técnicas fundamentales para el análisis de series temporales, especialmente para datos con tendencias y estacionalidades. Mientras que el SES es adecuado para datos sin una estructura compleja, el modelo Holt-Winters captura la tendencia y estacionalidad de manera efectiva. Ambas técnicas se evalúan usando métricas como el **MSE** y el **MAE** para determinar su precisión y eficacia en la predicción de datos futuros.

<H1>Proyecto: Modelos Estadísticos Básicos: Promedios Móviles y Exponenciales</H1>

**Paso 1: Configuración del entorno**

Primero, instala las bibliotecas necesarias. Si no las tienes instaladas, puedes hacerlo con:

```python
#pip install pandas numpy statsmodels matplotlib scikit-learn
```

**Paso 2: Importar las bibliotecas**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

**Paso 3: Cargar los datos**

Puedes cargar un conjunto de datos financieros o de ventas. Para este ejemplo, vamos a usar un conjunto de datos de ventas sintético que simula las ventas mensuales de una empresa.

```python
# Cargar el conjunto de datos de AirPassengers
df = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv', 
                 parse_dates=['Month'], index_col='Month')

# Renombrar la columna para mayor claridad
df.columns = ['Pasajeros']

# Mostrar los primeros registros para verificar
print(df.head())
```

```python
df
```

**Paso 4: Visualización de los datos**
    
Es útil visualizar los datos para ver las tendencias y estacionalidades.

```python
# Gráfico de los pasajeros mensuales
df['Pasajeros'].plot(title='Pasajeros Mensuales (1949-1960)', figsize=(10, 6))
plt.ylabel('Número de Pasajeros')
plt.show()
```

**Paso 5: Aplicar Suavizamiento Exponencial Simple (SES)**

Aplicamos el modelo de suavizamiento exponencial simple a los datos.

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

# Aplicar el suavizamiento exponencial simple
ses_model = SimpleExpSmoothing(df['Pasajeros']).fit(smoothing_level=0.2, optimized=False)
df['SES'] = ses_model.fittedvalues

# Gráfico comparando los datos reales con SES
df[['Pasajeros', 'SES']].plot(title='Suavizamiento Exponencial Simple (SES)', figsize=(10, 6))
plt.show()
```

**Paso 6: Aplicar el modelo de Holt-Winters**

El método de Holt-Winters captura tanto la tendencia como la estacionalidad de los datos. Lo aplicamos con estacionalidad multiplicativa, ya que los datos presentan un comportamiento exponencial.

```python
# Aplicar Holt-Winters con estacionalidad multiplicativa
hw_model = ExponentialSmoothing(df['Pasajeros'], trend='add', seasonal='mul', seasonal_periods=12).fit()

# Predecir los valores ajustados
df['Holt-Winters'] = hw_model.fittedvalues

# Gráfico comparando las ventas reales con el modelo Holt-Winters
df[['Pasajeros', 'Holt-Winters']].plot(title='Modelo Holt-Winters', figsize=(10, 6))
plt.show()
```

**Paso 7: Evaluación de los Modelos**
    
Evaluamos ambos modelos (SES y Holt-Winters) utilizando el error cuadrático medio (MSE) y el error absoluto medio (MAE).

```python
# Calcular el MSE y MAE para el SES
mse_ses = mean_squared_error(df['Pasajeros'], df['SES'])
mae_ses = mean_absolute_error(df['Pasajeros'], df['SES'])

# Calcular el MSE y MAE para Holt-Winters
mse_hw = mean_squared_error(df['Pasajeros'], df['Holt-Winters'])
mae_hw = mean_absolute_error(df['Pasajeros'], df['Holt-Winters'])

# Mostrar los resultados
print(f'SES - MSE: {mse_ses:.2f}, MAE: {mae_ses:.2f}')
print(f'Holt-Winters - MSE: {mse_hw:.2f}, MAE: {mae_hw:.2f}')
```

**Paso 8: Predicción a Futuro**

Finalmente, predecimos el número de pasajeros para los próximos 12 meses utilizando el modelo Holt-Winters.

```python
# Predecir los próximos 12 meses con el modelo de Holt-Winters
future_predictions = hw_model.forecast(12)

# Crear un DataFrame para las predicciones
future_dates = pd.date_range(start=df.index[-1], periods=13, freq='M')[1:]
future_df = pd.DataFrame({'Fecha': future_dates, 'Predicción': future_predictions})
future_df.set_index('Fecha', inplace=True)

# Gráfico de predicciones futuras
df['Pasajeros'].plot(label='Pasajeros Reales', figsize=(10, 6))
future_df['Predicción'].plot(label='Predicciones Holt-Winters', color='red')
plt.legend()
plt.title('Predicciones a Futuro con Holt-Winters')
plt.show()
```
