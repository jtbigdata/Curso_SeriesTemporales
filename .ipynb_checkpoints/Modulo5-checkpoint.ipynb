<H1>Marco Teórico para el Ajuste de un Modelo ARIMA</H1>

### **1. Modelos Autoregresivos (AR), de Media Móvil (MA), ARMA y ARIMA**

**Modelos AR (Autoregresivos)**:

Un modelo autoregresivo de orden $ p (AR(p))$ se basa en la idea de que el valor actual de la serie temporal puede ser explicado como una combinación lineal de sus valores pasados y un término de error.

**Definiciones**:
- **Serie Temporal $Y_t$**: Una secuencia de datos medidos en intervalos regulares de tiempo.
- **Parámetros $\phi_1, \phi_2, \ldots, \phi_p $**: Coeficientes que determinan la influencia de los valores pasados de la serie temporal.
- **Término de Error $\epsilon_t $**: Un término aleatorio que representa la variabilidad no explicada por el modelo.

**Ecuación del Modelo AR(p)**:

$ Y_t = \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \cdots + \phi_p Y_{t-p} + \epsilon_t $

donde:
- $Y_t$ es el valor de la serie temporal en el tiempo $t$.
- $\phi_i$ son los coeficientes autoregresivos.
- $\epsilon_t$ es el término de error en el tiempo $t$.

**Modelos MA (Media Móvil)**:

Un modelo de media móvil de orden $q (MA(q))$ utiliza los errores pasados para modelar la serie temporal. Es útil cuando los valores pasados no explican bien la serie temporal, pero los errores sí.

**Definiciones**:
- **Parámetros $ \theta_1, \theta_2, \ldots, \theta_q $**: Coeficientes que determinan la influencia de los errores pasados en el valor actual.

**Ecuación del Modelo MA(q)**:

$ Y_t = \epsilon_t + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + \cdots + \theta_q \epsilon_{t-q} $

donde:
- $ \theta_i $ son los coeficientes de media móvil.

**Modelos ARMA (Autoregresivos y de Media Móvil)**:

Un modelo ARMA de orden $(p, q)$  combina los componentes $AR(p)$ y $MA(q)$ para modelar una serie temporal.

**Ecuación del Modelo ARMA(p, q)**:

$ Y_t = \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \cdots + \phi_p Y_{t-p} + \epsilon_t + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + \cdots + \theta_q \epsilon_{t-q} $

**Modelos ARIMA (Autoregresivos Integrados de Media Móvil)**:

Un modelo ARIMA de orden $(p, d, q)$ extiende el modelo ARMA para manejar series temporales no estacionarias mediante la diferenciación.

**Definiciones**:
- **Diferenciación $d$**: Proceso de transformar la serie temporal para hacerla estacionaria. La diferenciación de orden $d$ se refiere a aplicar el operador de diferenciación $d$ veces.

**Ecuación del Modelo ARIMA(p, d, q)**:

Para una serie diferenciada, el modelo ARIMA se define como:

$ \Delta^d Y_t = \phi_1 \Delta^d Y_{t-1} + \phi_2 \Delta^d Y_{t-2} + \cdots + \phi_p \Delta^d Y_{t-p} + \epsilon_t + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + \cdots + \theta_q \epsilon_{t-q} $

donde:
- $ \Delta^d $ es el operador de diferenciación aplicado $d$  veces.

### **2. Identificación de Estacionariedad**

**Estacionariedad**: Es una propiedad de una serie temporal donde sus características estadísticas, como la media y la varianza, no cambian con el tiempo. Las series estacionarias son más fáciles de modelar.

**Prueba de Dickey-Fuller**:

La prueba de Dickey-Fuller se utiliza para verificar si una serie temporal tiene una raíz unitaria, lo que indicaría que no es estacionaria. La hipótesis nula es que la serie tiene una raíz unitaria.

**Ecuación del Modelo de Regresión de Dickey-Fuller**:

$\Delta Y_t = \alpha + \beta t + \rho Y_{t-1} + \epsilon_t $

donde:
- $ \Delta Y_t $ es la diferencia de la serie temporal.
- $ \alpha $ es una constante.
- $ \beta t $ es un término de tendencia.
- $ \rho $ es el coeficiente del rezago.
- $ \epsilon_t $ es el término de error.

**Hipótesis**:
- **Hipótesis Nula $ H_0$**: La serie tiene una raíz unitaria (no es estacionaria).
- **Hipótesis Alternativa $ H_1$**: La serie es estacionaria.

### **3. Selección de Parámetros para ARIMA**

**Gráficos ACF y PACF**:

- **ACF (Autocorrelation Function)**: Mide la correlación entre una serie temporal y sus rezagos. Se utiliza para determinar el orden del componente MA.

- **PACF (Partial Autocorrelation Function)**: Mide la correlación parcial entre una serie temporal y sus rezagos. Se utiliza para determinar el orden del componente AR.

**Interpretación**:
- **ACF**: Se observa para encontrar el valor de $q$. El primer lag donde la ACF se corta indica el orden del componente MA.
- **PACF**: Se observa para encontrar el valor de $p$ . El primer lag donde la PACF se corta indica el orden del componente AR.

### **4. Ajuste del Modelo ARIMA**

**Ajuste del Modelo**:

El modelo ARIMA se ajusta a la serie temporal diferenciada utilizando los parámetros $p$, $d$ y $q$. Se estima el modelo para obtener los parámetros autoregresivos y de media móvil.

**Ecuación del Modelo ARIMA Ajustado**:

Para una serie diferenciada:

$ \Delta^d Y_t = \phi_1 \Delta^d Y_{t-1} + \phi_2 \Delta^d Y_{t-2} + \cdots + \phi_p \Delta^d Y_{t-p} + \epsilon_t + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + \cdots + \theta_q \epsilon_{t-q} $

donde:
- $ \phi_i $ son los parámetros autoregresivos estimados.
- $ \theta_i $ son los parámetros de media móvil estimados.

### **5. Validación del Modelo**

**Errores de Predicción**:

- **MSE (Mean Squared Error)**:

$ \text{MSE} = \frac{1}{n} \sum_{t=1}^n (Y_t - \hat{Y}_t)^2 $

donde $ Y_t $ es el valor real y $\hat{Y}_t $ es el valor predicho por el modelo.

- **RMSE (Root Mean Squared Error)**:

$ \text{RMSE} = \sqrt{\text{MSE}} $

Estos errores miden la precisión del modelo al hacer predicciones. El MSE mide el promedio de los errores al cuadrado, mientras que el RMSE proporciona una medida en las mismas unidades que la serie temporal.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python

url='/home/julio/jupyter_files/Curso_SeriesTemporales/retail_sales_dataset.csv'
df=pd.read_csv(url)
```

```python
# Mostrar información y primeras filas del dataframe
print(df.info())
```

```python
print(df.head())
```

```python
# Visualizar la serie temporal
df['Total Amount'].plot(figsize=(12, 6))
plt.title('Serie Temporal del Monto Total')
plt.xlabel('Fecha')
plt.ylabel('Monto Total')
plt.show()
```

**Paso 3: Verificar Estacionariedad**

```python
def test_stationarity(timeseries):
    result = adfuller(timeseries, autolag='AIC')
    print('Estadístico de prueba:', result[0])
    print('Valor p:', result[1])
    print('Número de retardos utilizados:', result[2])
    print('Número de observaciones utilizadas:', result[3])
    print('Valores críticos:')
    for key, value in result[4].items():
        print(f'\t{key}: {value}')

# Aplicar la prueba a la serie temporal
test_stationarity(df['Total Amount'])
```

1. **Estadístico de prueba**: El valor del estadístico de prueba es **-31.7760**. Este valor se compara con los valores críticos (al 1%, 5% y 10%) para determinar si se puede rechazar la hipótesis nula (la hipótesis nula asume que la serie temporal **no es estacionaria**).

2. **Valor p**: El valor p es **0.0**, lo cual es extremadamente bajo. Esto indica que hay suficiente evidencia para **rechazar la hipótesis nula** de que la serie temporal no es estacionaria.

3. **Número de retardos utilizados**: Se utilizaron **0 retardos** en esta prueba.

4. **Número de observaciones utilizadas**: La prueba se realizó con **999 observaciones**.

5. **Valores críticos**: Los valores críticos a los niveles de significancia del 1%, 5%, y 10% son:
   - 1%: **-3.4369**
   - 5%: **-2.8644**
   - 10%: **-2.5683**

### Conclusión:
Dado que el estadístico de prueba (-31.7760) es mucho menor que el valor crítico al 1% (-3.4369), y el valor p es prácticamente 0, podemos **rechazar con confianza la hipótesis nula**. Esto sugiere que la serie temporal **es estacionaria**.

En resumen, si querías verificar si la serie de tiempo `Total Amount` era estacionaria, los resultados indican que **sí lo es**, ya que no tiene una raíz unitaria.

Si la serie no es estacionaria, aplicaremos diferenciación.

```python
# Diferenciación
df['Total Amount_diff'] = df['Total Amount'] - df['Total Amount'].shift(1)
df.dropna(inplace=True)

# Graficar la serie diferenciada
df['Total Amount_diff'].plot(figsize=(12, 6))
plt.title('Serie Temporal del Monto Total Diferenciada')
plt.xlabel('Fecha')
plt.ylabel('Monto Total Diferenciado')
plt.show()

# Verificar estacionariedad después de la diferenciación
test_stationarity(df['Total Amount_diff'])
```

1. **Diferenciación**:
   - El código diferencia la serie temporal original (`Total Amount`) restando cada valor de su valor anterior (es decir, `shift(1)`), para eliminar tendencias o patrones de no estacionariedad.
   - La nueva serie resultante se almacena en la columna `Total Amount_diff`.
   - Luego se eliminan las filas con valores nulos resultantes de la diferenciación (primera fila), usando `dropna()`.

2. **Gráfico de la serie diferenciada**:
   - Se genera un gráfico para visualizar cómo se ve la serie temporal después de la diferenciación. Esto puede ayudar a observar si el comportamiento general ha cambiado después de eliminar la tendencia.

3. **Prueba de estacionariedad**:
   - Se realiza la prueba ADF en la serie diferenciada (`Total Amount_diff`) para verificar si ahora es estacionaria.

**Resultado de la prueba ADF en la serie diferenciada:**

- **Estadístico de prueba**: **-12.6438**. Un valor mucho menor que los valores críticos, lo que indica que es muy probable que la serie sea estacionaria.
  
- **Valor p**: **1.4162e-23** (un valor extremadamente pequeño), lo que proporciona una fuerte evidencia para **rechazar la hipótesis nula** de no estacionariedad.
  
- **Número de retardos utilizados**: **20 retardos** fueron seleccionados automáticamente para optimizar el ajuste de la prueba.
  
- **Número de observaciones utilizadas**: **978 observaciones** se incluyeron en la prueba.

- **Valores críticos**:
   - 1%: **-3.4371**
   - 5%: **-2.8645**
   - 10%: **-2.5683**

**Conclusión:**
El **estadístico de prueba** (-12.6438) es mucho más bajo que el valor crítico al 1% (-3.4371), lo que indica que podemos **rechazar la hipótesis nula** de que la serie temporal no es estacionaria. Dado que el valor p es casi cero, hay suficiente evidencia para concluir que, después de la diferenciación, la serie temporal `Total Amount_diff` **es estacionaria**.

Por lo tanto, la diferenciación ha sido efectiva para transformar la serie y hacerla más apropiada para modelar procesos estacionarios.

**Paso 4: Selección de Parámetros (p, d, q)**

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Graficar ACF y PACF
plt.figure(figsize=(12, 6))

plt.subplot(121)
plot_acf(df['Total Amount_diff'], lags=40, ax=plt.gca())
plt.title('ACF')

plt.subplot(122)
plot_pacf(df['Total Amount_diff'], lags=40, ax=plt.gca())
plt.title('PACF')

plt.show()
```

**Paso 5: Ajustar el Modelo ARIMA**

```python
# Definir los parámetros (ajusta estos valores basados en los gráficos ACF y PACF)
p, d, q = 5, 1, 0

# Dividir los datos en entrenamiento y prueba
train = df['Total Amount'][:int(len(df)*0.8)]
test = df['Total Amount'][int(len(df)*0.8):]

# Ajustar el modelo ARIMA
model = ARIMA(train, order=(p, d, q))
model_fit = model.fit()
```

```python
# Resumen del modelo
print(model_fit.summary())
```

El output muestra los resultados del ajuste de un modelo **ARIMA(5, 1, 0)** a los datos de entrenamiento (`Total Amount`), que fue entrenado con el 80% de los datos. Vamos a desglosar los elementos más importantes del resultado:

1. **Resumen del modelo ARIMA(5, 1, 0)**:
   - El modelo ARIMA tiene 5 términos autorregresivos (AR), 1 diferenciación (para hacer la serie estacionaria) y 0 términos de media móvil (MA).
   - Se utilizaron **799 observaciones** para el ajuste del modelo.

2. **Log Likelihood (Verosimilitud Logarítmica)**: 
   - El valor de **-6272.663**. La verosimilitud logarítmica indica qué tan bien el modelo se ajusta a los datos. Un valor mayor (menos negativo) es mejor.

3. **Criterios de Información**:
   - **AIC (Akaike Information Criterion)**: **12557.325**. Este criterio se utiliza para evaluar la calidad del modelo, penalizando la complejidad. Un valor más bajo indica un mejor modelo.
   - **BIC (Bayesian Information Criterion)**: **12585.418**. Similar al AIC, pero con una penalización más fuerte para modelos más complejos.
   - **HQIC (Hannan-Quinn Information Criterion)**: **12568.119**. Un criterio similar pero con una penalización menos agresiva que el BIC.

4. **Coeficientes AR (Autorregresivos)**:
   - **ar.L1**: Coeficiente de autorregresión en el retardo 1, **-0.8233**, con un valor p de 0.000, lo que indica que es estadísticamente significativo.
   - **ar.L2**: Coeficiente en el retardo 2, **-0.6347**, también significativo.
   - **ar.L3**: Coeficiente en el retardo 3, **-0.4804**, significativo.
   - **ar.L4**: Coeficiente en el retardo 4, **-0.3197**, significativo.
   - **ar.L5**: Coeficiente en el retardo 5, **-0.1491**, significativo.
   - Todos los coeficientes AR son negativos, lo que indica que los valores pasados están inversamente correlacionados con los valores futuros.

5. **Sigma^2** (Varianza del error residual): 
   - **3.948e+05**. Este valor indica la varianza de los errores del modelo. Un valor más bajo es mejor, ya que implica que el modelo tiene menores errores de predicción.

6. **Estadísticas de bondad de ajuste**:
   - **Ljung-Box (Q)**: **0.20** con una probabilidad (p-value) de **0.66**. Esto sugiere que no hay autocorrelación significativa en los residuos del modelo, lo cual es bueno.
   - **Jarque-Bera (JB)**: **140.30** con un valor p de **0.00**. Esto indica que los residuos no siguen una distribución normal, lo cual podría ser un indicio de que el modelo podría beneficiarse de mejoras o transformaciones adicionales.
   - **Heterocedasticidad (H)**: **1.00**, lo cual indica que los errores tienen varianza constante, lo que es deseable.

7. **Skewness (Asimetría)**: 
   - **1.01**. Los residuos están algo asimétricos, lo que podría influir en la normalidad de los mismos.

8. **Kurtosis**: 
   - **3.40**. Indica que los residuos tienen una leve "cola pesada" en comparación con una distribución normal, lo cual también puede influir en la precisión de las predicciones.

**Conclusión:**
El modelo **ARIMA(5, 1, 0)** ha sido ajustado correctamente y tiene coeficientes significativos. Aunque el ajuste del modelo parece razonable, algunos resultados como la prueba de Jarque-Bera indican que los residuos no son perfectamente normales, lo cual podría afectar las predicciones futuras. Sin embargo, la ausencia de autocorrelación significativa en los residuos (Ljung-Box) sugiere que el modelo captura bien la dinámica de los datos.

Si observas que las predicciones no son precisas, puedes considerar ajustar los valores de los parámetros **p** y **q** o explorar modelos SARIMA si sospechas estacionalidad.

**Paso 6: Predicción y Evaluación**

```python
# Realizar predicciones
predictions = model_fit.forecast(steps=len(test))

# Graficar resultados
plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label='Entrenamiento')
plt.plot(test.index, test, label='Real')
plt.plot(test.index, predictions, label='Predicción', color='red')
plt.title('Predicción del Monto Total')
plt.xlabel('Fecha')
plt.ylabel('Monto Total')
plt.legend()
plt.show()
```

```python
# Evaluar el modelo
mse = mean_squared_error(test, predictions)
rmse = np.sqrt(mse)

print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
```
