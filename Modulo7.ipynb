<H1>Marco Teorico</H1>



**1. Introducción a la Volatilidad en Series Temporales**

La **volatilidad** se refiere a la variabilidad o dispersión de los precios de un activo financiero a lo largo del tiempo. En series temporales financieras, la volatilidad no es constante y tiende a cambiar con el tiempo. Por lo tanto, es importante modelar esta variabilidad para evaluar el riesgo y la incertidumbre en los mercados financieros.

**1.1 Volatilidad Condicional**

La volatilidad condicional es la medida de la variabilidad futura de los precios, condicionada a la información disponible hasta el momento. La volatilidad condicional se modela a menudo mediante modelos que capturan la heterocedasticidad, es decir, la variabilidad de la volatilidad en el tiempo.

**2. Modelos GARCH (Generalized Autoregressive Conditional Heteroskedasticity)**

El modelo GARCH es una extensión del modelo ARCH (Autoregressive Conditional Heteroskedasticity) que permite modelar la volatilidad condicional de manera más flexible. Los modelos GARCH se utilizan para capturar la dependencia de la volatilidad en el tiempo.

**2.1 Modelo ARCH (Autoregressive Conditional Heteroskedasticity)**

El modelo ARCH fue propuesto por Robert Engle en 1982. La fórmula del modelo ARCH(p) es:

$ \sigma_t^2 = \alpha_0 + \alpha_1 \varepsilon_{t-1}^2 + \alpha_2 \varepsilon_{t-2}^2 + \cdots + \alpha_p \varepsilon_{t-p}^2 $

donde:
- $\sigma_t^2$ es la varianza condicional en el tiempo $t$,
- $\alpha_0$ es el término constante,
- $\alpha_i$ $(para (i = 1, \ldots, p))$ son los coeficientes de los rezagos de los errores cuadráticos,
- $\varepsilon_{t-i}$ son los residuos del modelo de media en el tiempo $t-i$.

#### **2.2 Modelo GARCH (Generalized Autoregressive Conditional Heteroskedasticity)**

El modelo GARCH fue propuesto por Tim Bollerslev en 1986. Es una generalización del modelo ARCH y combina tanto la dependencia de los errores pasados como la dependencia de la volatilidad pasada. La fórmula del modelo GARCH(p, q) es:

$ \sigma_t^2 = \alpha_0 + \sum_{i=1}^q \alpha_i \varepsilon_{t-i}^2 + \sum_{j=1}^p \beta_j \sigma_{t-j}^2 $

donde:
- $\sigma_t^2$ es la varianza condicional en el tiempo $t$,
- $\alpha_0$ es el término constante,
- $\alpha_i$ (para $i = 1, \ldots, q$) son los coeficientes de los rezagos de los errores cuadráticos,
- $\beta_j$ (para $j = 1, \ldots, p$) son los coeficientes de los rezagos de la varianza condicional pasada,
- $\varepsilon_{t-i}$ son los residuos del modelo de media en el tiempo $t-i$,
- $\sigma_{t-j}^2$ son las varianzas condicionales pasadas.

**2.3 Interpretación del Modelo GARCH**

- **$\alpha_0$**: Representa la varianza incondicional, el nivel base de volatilidad.
- **$\alpha_i$**: Captura el impacto de los choques pasados en la volatilidad actual. Un mayor valor de $\alpha_i$ indica que los choques pasados tienen un impacto significativo en la volatilidad actual.
- **$\beta_j$**: Representa el efecto de la volatilidad pasada en la volatilidad actual. Un mayor valor de $\beta_j$ indica una alta persistencia en la volatilidad.

**3. Aplicaciones de GARCH en Series Temporales Financieras**

Los modelos GARCH se aplican ampliamente en finanzas para:

- **Predecir la volatilidad futura**: Proporcionan estimaciones de la variabilidad futura de los precios de los activos, lo que es útil para la gestión del riesgo y la toma de decisiones de inversión.
- **Análisis de riesgo**: Ayudan a evaluar el riesgo asociado con los activos financieros, permitiendo a los inversores ajustar sus estrategias para mitigar el riesgo.
- **Valuación de derivados**: Los modelos de volatilidad, como GARCH, se utilizan para valorar opciones y otros instrumentos financieros derivados que dependen de la volatilidad del activo subyacente.

**1. Introducción a la Volatilidad y el Modelo GARCH**

La volatilidad mide la variabilidad de los precios de un activo a lo largo del tiempo, lo que es clave en el análisis de riesgo financiero. La volatilidad no es constante, tiende a cambiar a lo largo del tiempo, y aquí es donde los modelos GARCH (Generalized Autoregressive Conditional Heteroskedasticity) son útiles.

El modelo GARCH ayuda a modelar la heterocedasticidad condicional, es decir, cómo la volatilidad cambia con el tiempo.

**2. Paso a Paso: Implementación del Modelo GARCH**

**Paso 1: Instalación de librerías necesarias**

**Paso 2: Importación de librerías**

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from arch import arch_model
```

**Paso 3: Descarga de datos financieros**

Vamos a utilizar datos históricos de precios de Bitcoin (BTC) como ejemplo. Para obtener los datos, utilizaremos yfinance.

```python
# Descargar datos de Bitcoin (BTC-USD) desde Yahoo Finance
data = yf.download("BTC-USD", start="2020-01-01", end="2023-01-01")
data.head()
```

**Paso 4: Cálculo de los rendimientos diarios**

Para aplicar el modelo GARCH, necesitamos los rendimientos logarítmicos en lugar de los precios crudos.

```python
# Cálculo de los rendimientos logarítmicos
data['Log_Returns'] = np.log(data['Close'] / data['Close'].shift(1))
data = data.dropna()  # Eliminar los valores nulos
data['Log_Returns'].plot(title="Rendimientos Logarítmicos de BTC", figsize=(10, 6))
plt.show()
```

**Paso 5: Modelo GARCH**

El modelo GARCH(1,1) es el más utilizado en la práctica. Se refiere a que tomamos en cuenta un término de media autoregresiva y uno de heterocedasticidad condicional.

```python
# Definir el modelo GARCH(1,1)
model = arch_model(data['Log_Returns'], vol='Garch', p=1, q=1)

# Ajustar el modelo a los datos
garch_fit = model.fit(disp="off")
print(garch_fit.summary())
```

Este resultado representa los resultados de ajustar un modelo **GARCH** (Generalized Autoregressive Conditional Heteroskedasticity) sobre los **rendimientos logarítmicos** de una serie temporal financiera. Aquí tienes un desglose en detalle:

### 1. **Modelo de Media (Media Constante)**
- **Variable Dependiente (Log_Returns)**: La variable que se está modelando son los rendimientos logarítmicos de un activo financiero o índice.
- **Modelo de Media: Media Constante**: Asume que la media de los rendimientos es constante a lo largo del tiempo.
- **R-cuadrado (0.000)**: Indica que el modelo de media constante no explica ninguna variación en los rendimientos logarítmicos (esto es esperable en este tipo de modelo).
- **Log-Likelihood (2029.50)**: Medida del ajuste del modelo; valores más altos indican un mejor ajuste.
- **AIC (-4050.99)** y **BIC (-4031.00)**: Criterios de información para comparar modelos, donde valores más bajos son preferibles.

#### Coeficientes del Modelo de Media:
- **mu (1.5843e-03)**: El valor constante de la media es aproximadamente 0.0016, pero con un **p-valor de 0.150** (no es estadísticamente significativo al nivel del 5%), lo que indica que no podemos rechazar la hipótesis de que la media verdadera sea cero.

### 2. **Modelo de Volatilidad (GARCH)**
- Este modelo estima la volatilidad (varianza) de los rendimientos logarítmicos a lo largo del tiempo, teniendo en cuenta tanto los rendimientos pasados como la volatilidad pasada.

#### Coeficientes del Modelo de Volatilidad:
- **omega (1.5990e-04)**: El término constante en la ecuación de la varianza. Su **p-valor (0.142)** indica que no es estadísticamente significativo.
- **alpha[1] (0.0572)**: Mide la influencia de los choques recientes (rendimientos pasados al cuadrado) sobre la volatilidad actual. El **p-valor (0.038)** significa que es significativo, lo que indica que los choques recientes impactan la volatilidad actual.
- **beta[1] (0.8436)**: Mide la persistencia de la volatilidad. Un **valor alto (0.8436)** y un **p-valor muy significativo (<0.0001)** indican que la volatilidad pasada tiene un fuerte efecto sobre la volatilidad actual.

### 3. **Conclusión**
- El **rendimiento medio (mu)** no es estadísticamente significativo, lo que implica que la media constante está probablemente cerca de cero.
- El **modelo GARCH** muestra una dinámica significativa en la volatilidad, donde tanto los choques recientes como la volatilidad pasada influyen en la volatilidad actual. El valor alto de **beta (0.8436)** sugiere que la volatilidad es muy persistente, es decir, una vez que aumenta, tiende a mantenerse alta durante un tiempo.

**Paso 6: Predicción de la Volatilidad**

Con el modelo ajustado, podemos predecir la volatilidad futura.

```python
# Predicción de volatilidad para los próximos 10 días
forecast = garch_fit.forecast(horizon=10)

# Extraer la volatilidad condicional predicha
volatility = np.sqrt(forecast.variance.iloc[-1])
print("Volatilidad pronosticada para los próximos 10 días:")
print(volatility)
```

**Paso 7: Visualización de la Volatilidad**

Visualizamos la volatilidad condicional a lo largo del tiempo.

```python
# Volatilidad condicional estimada
data['Volatility'] = garch_fit.conditional_volatility
data[['Log_Returns', 'Volatility']].plot(subplots=True, figsize=(10, 8))
plt.show()
```
